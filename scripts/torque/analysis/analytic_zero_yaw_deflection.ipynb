```python
import matplotlib.pyplot as plt
import numpy as np

plt.style.use("torque.mplstyle")

%matplotlib widget
```

```python
psi = np.linspace(0, 2 * np.pi, 1000)
psi_deg = np.rad2deg(psi)
A0 = 5
Atiltnew = 3
Atilt = 1
Ayaw = -2.5
baseline = A0 + Atilt * np.cos(psi) + Ayaw * np.sin(psi)
free_yaw = A0 + Atiltnew * np.cos(psi) + Ayaw * np.sin(psi)
zero_yaw = A0 + Atiltnew * np.cos(psi)

default_fig_width = plt.rcParams["figure.figsize"][0]
fig, ax = plt.subplots(figsize=(default_fig_width / 2, 2.5))

ax.plot(psi_deg, baseline, label="Baseline")
ax.plot(psi_deg, free_yaw, label="Free yaw")
ax.plot(psi_deg, zero_yaw, label="Zero yaw")

# ax.axvline(180, color="gray", linestyle="--")
# ax.axhline(2, color="gray", linestyle="--")
ax.scatter(180, 2, color="gray", label="Reference")

ax.set_xlabel("Azimuth angle (deg)")
ax.set_ylabel("Tip deflection (m)")
ax.legend(loc="upper right")
ax.grid()

ax.set_xticks(np.arange(0, 361, 60))

# ax.set_ylim((0, 14))

plt.tight_layout()
plt.savefig("../figures/analytic_deflection_rotating.pdf")
```

```python
fig, ax = plt.subplots(subplot_kw={"projection": "polar"})

ax.plot(psi, baseline, label="Baseline", linewidth=2)
ax.plot(psi, free_yaw, label="Free yaw", linewidth=2)
ax.plot(psi, zero_yaw, label="Zero yaw deflection", linewidth=2)

# Mark 0 degrees (tower position)
ax.axvline(0, color="gray", linestyle="--", alpha=0.5, linewidth=1)

ax.legend(loc="upper right", bbox_to_anchor=(1.3, 1.1))
ax.set_theta_zero_location("N")  # 0 degrees at top
ax.set_theta_direction(-1)  # Clockwise

plt.tight_layout()
# plt.savefig("figures/zero_yaw_deflection_polar.pdf")
```
